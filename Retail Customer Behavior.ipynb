```python
df['Review Rating'] = df.groupby('Category') ['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ","_")
df = df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
```

```python
df.head()
```

```python
bins = [0, 25, 45, 65, 101]
labels = ['young', 'adult', 'middle age', 'senior']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)
display(df.head())
```

```python
cols = df.columns.tolist()
age_group_col = cols.pop(cols.index('age_group'))
age_index = cols.index('age')
cols.insert(age_index + 1, age_group_col)
df = df[cols]
display(df.head())
```

```python
frequency_mapping = {
    'Every 3 Months': 90,
    'Annually': 365,
    'Quarterly': 90,
    'Fortnightly': 14,
    'Once per month': 30,
    'Weekly': 7,
    'Bi-Weekly': 14,
    'Daily': 1
}

df['frequency_of_purchases_days'] = df['frequency_of_purchases'].map(frequency_mapping)

display(df[['frequency_of_purchases', 'frequency_of_purchases_days']].head())
```

### Checking Redundancy between 'discount_applied' and 'promo_code_used'

First, let's look at the unique values and their counts for both columns.

```python
print("Discount Applied Value Counts:")
print(df['discount_applied'].value_counts())
print("\nPromo Code Used Value Counts:")
print(df['promo_code_used'].value_counts())
```

Now, let's create a crosstabulation to see the relationship between these two columns.

```python
contingency_table = pd.crosstab(df['discount_applied'], df['promo_code_used'])
print("\nContingency Table (Discount Applied vs. Promo Code Used):")
display(contingency_table)
```

```python
df = df.drop(columns=['promo_code_used'])
display(df.head())
```

### Saving DataFrame to BigQuery

Now that you have processed your data and have a BigQuery connection, you can save your DataFrame `df` to a new BigQuery table for further analysis. You'll need to specify a dataset name (it will be created if it doesn't exist) and a table name.

```python
import pandas_gbq

# Define your dataset and table names
dataset_name = 'customer_data'  # Replace with your desired dataset name
table_name = 'customer_shopping_behavior_processed' # Replace with your desired table name
project_id = "jpmc-portfolio-project"

# Construct the full table ID
table_id = f"{project_id}.{dataset_name}.{table_name}"

try:
    # Save the DataFrame to BigQuery
    pandas_gbq.to_gbq(df, table_id, project_id=project_id, if_exists='replace')
    print(f"DataFrame successfully saved to BigQuery table: {table_id}")
except Exception as e:
    print(f"Error saving DataFrame to BigQuery: {e}")
```
